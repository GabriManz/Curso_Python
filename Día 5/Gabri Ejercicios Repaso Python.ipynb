## Ejercicio 1

Contar cuántos valores hay en una lista de enteros. Contar solo hasta el número 0.

```python
nums = [6, 8, 5, 2, 0, 6, 5]
contador = 0

for num in nums:
    contador = contador + 1
    if num == 0:
        break


print(contador)
```

## Ejercicio 3

Calcular el valor medio de los valores que hay en una lista de enteros. Calcular solo hasta el número 0.

```python
nums = [6,8,5,2,0,6,5]
suma = 0
contador = 0
#Respuesta:
#Debe dar 4.2

for num in nums:
    suma = num + suma
    contador = contador + 1
    if num == 0:
        break

media = suma/contador
print(media)
```

## Ejercicio 5 (Opcional)

Crear una lista a partir de una secuencia de valores de entrada terminada en *.
Una vez creada, modificarla elevando al cuadrado todos los pares y al cubo todos los impares, y mostrarla.

```python
lista = [3,56,6,7,"*",2,78]
new_lista = []
#Respuesta:
#Debe dar [27, 3136, 36, 343]

for num in lista:
  if num == '*':
    break
  elif num%2 == 0:
    new_lista.append(num**2)
  elif num%2 != 0:
    new_lista.append(num**3)

print(new_lista)
```
